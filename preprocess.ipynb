```python
import os

# Define the directory containing the files
directory = 'C:/Users/haZAR/Desktop/New folder/train/labels'

# Function to map the values as specified
def map_value(value):
    if value == '1':
        return '0'
    elif value == '2':
        return '1'
    elif value == '3':
        return '2'
    elif value == '4':
        return '3'
    else:
        return value

# List all files in the directory
file_paths = os.listdir(directory)

# Process each file
for file_name in file_paths:
    file_path = os.path.join(directory, file_name)
    
    # Read the file content
    with open(file_path, 'r') as f:
        lines = f.readlines()
    
    # Write the modified content back to the file
    with open(file_path, 'w') as f:
        for line in lines:
            parts = line.strip().split()
            if parts:
                parts[0] = map_value(parts[0])
            f.write(' '.join(parts) + '\n')
```

```python
import os
import json

file_path = "C:/Users/haZAR/Desktop/New folder/hello"
file_paths = os.listdir(file_path)

for k in file_paths:
    path = str(k)
    # Open the JSON file
    with open(os.path.join(file_path, path)) as file:
        read_file = json.load(file)
        instances = read_file["instances"]
        
        for i in instances:
            class_id = i["classId"]
            points = i["points"]
            x1 = points['x1']
            x2 = points['x2']
            y1 = points['y1']
            y2 = points['y2']
            
            # Calculate normalized coordinates
            centre_x = abs(x1 + x2) / (2 * 640)
            centre_y = abs(y1 + y2) / (2 * 480)
            width = abs(x2 - x1) / 640
            height = abs(y2 - y1) / 480
            
            # Extract the file name without the extension
            file_strip = path.strip().split('/')
            last_part = file_strip[-1]
            name = last_part.replace('.png.json', '')
            
            # Write the data to a text file
            with open(name + '.txt', 'a') as f:
                f.write(f"{class_id} {centre_x} {centre_y} {width} {height}\n")
```
